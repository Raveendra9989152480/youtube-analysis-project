```python
c=widgets.Dropdown(
      options=video['Genre'].unique(),
      value='Music',
      description='Genre',
      disabled=False
)
def graph(a):
  dr=video[video['Genre']==a]
  gt=dr.groupby('channelTitle')
  gt=gt.sum().sort_values(by='view_count',ascending=False).head()
  sns.barplot(y=gt.index,x=gt['view_count'])
  plt.title(a)
  plt.xlabel('views')
  plt.ylabel('channel names')

  plt.show()
```

```python
widgets.interact(graph,a=c)
```

```python
k1=widgets.Dropdown(
    options=video['Genre'].unique(),
    value='Music',
    description='Genre',
    disabled=False
)
def genre(m):
  gen=video[video['Genre']==m]
  k2=widgets.Dropdown(
  options=gen['channelTitle'].unique(),
  value=gen['channelTitle'].unique()[0],
  description='channel',
  disabled=False
  )
  def channel(d):
    ht=gen.sort_values(by='view_count',ascending=False).head()
    sns.barplot(y='title_x',x='view_count',data=ht)
    plt.show()
  widgets.interact(channel,d=k2)
widgets.interact(genre,m=k1)

```

```python
k2=widgets.Dropdown(
    options=video['channelTitle'].unique(),
    value=video['channelTitle'].unique()[0],
    description='channelTitle',
    disabled=False
)
def channel(k):
  gen=video[video['channelTitle']==k]
  k3=widgets.Dropdown(
      options=gen['title_x'].unique(),
      value=gen['title_x'].unique()[0],
      description='video title',
      disabled=False
  )
  def vid(d):
    ht=gen.sort_values(by='view_count',ascending=False).head()
    sns.lineplot(y='trending_date',x='view_count',data=ht)
    plt.show()
  widgets.interact(vid,d=k3)
widgets.interact(channel,k=k2)
```

```python
ct=widgets.Dropdown(
    options=video['channelTitle'].unique(),
    value=video['channelTitle'].unique()[0],
    description='Channel Title :',
    disabled=False
  )
def channel(d):
  cl=video[video['channelTitle']==d]
  vd=widgets.Dropdown(
      options=cl['title_x'].unique(),
      value=cl['title_x'].unique()[0],
      description='Video Title',
      disabled=False
      )
  def vid(c):
    vt=cl.sort_values('view_count',ascending=False)
    sns.lineplot(x='trending_date',y='view_count',data=vt)
    plt.show()
  widgets.interact(vid,c=vd)
widgets.interact(channel,d=ct)
```
